```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Run the magic command to activate the ipympl backend
%matplotlib widget
```

```python
# Create grid of plots
fig, axes = plt.subplots(2, 2);
# Add plots to our Figure
axes[0,0].hist(np.random.standard_normal(100), bins=20, color="black", alpha=0.3);
axes[0,1].scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30));
axes[1,0].plot(np.random.standard_normal(50).cumsum(), color="black",linestyle="dashed");
```

```python
# Remove padding
fig.subplots_adjust(wspace=0, hspace=0)
```

```python
# Dashed plot with markers and pink line
x = np.arange(30)
y = np.arange(30) + 5 * np.random.standard_normal(30)
fig, ax = plt.subplots();
ax.plot(x, y, linestyle="dashdot", color="g");
```

```python
# Standard plot
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum());
```

```python
# Set tick locations
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
ticks_y = ax.set_yticks([-10, 0, 10, 20])
```

```python
# Set tick labels for those tick points
labels = ax.set_xticklabels(["dog", "fish", "cat", "sheep", "fish"], rotation=30, fontsize=10)
```

```python
# Add an axis label (title)
ax.set_xlabel("People of a kind");
ax.set_ylabel("Growth (years)");
```

```python
fig.subplots_adjust(wspace=5, hspace=5)
```

```python
# Give our plot a title
ax.set_title("Change in animal population");
```

**Adding a legend**

```python
fig, ax = plt.subplots();
ax.plot(np.random.randn(1000).cumsum(), color="black", label="one");
ax.plot(np.random.randn(1000).cumsum(), color="green", linestyle="dashed",label="two");
ax.plot(np.random.randn(1000).cumsum(), color="magenta", linestyle="dotted",label="three");
ax.legend();
```

```python
# Adding text on the plot
ax.text(0, -20, "Bottom corner of my plot", family="monospace", fontsize=13);
```

```python
# Shapes
fig, ax = plt.subplots();
rect = plt.Rectangle((0.2, 0.75), 0.7, 0.35, color="black", alpha=1);
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3);
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],color="green", alpha=0.5);
ax.add_patch(rect);
ax.add_patch(circ);
ax.add_patch(pgon);
```
