**3. DataFrames**

```python
import numpy as np
import pandas as pd
```

```python
# Data as dictionary
data = {"province": ["Ontario", "Ontario", "Ontario", "Alberta", "Alberta", "Alberta"],
"year": [2010, 2011, 2012, 2011, 2012, 2013],
"pop": [13.1, 12.9, 14.6, 3.6, 4.3, 4.0]}
#data
frame = pd.DataFrame(data)
frame
```

```python
# Head and tail
#print(frame.head())
frame.tail()
```

```python
# Specify column ordering
pd.DataFrame(data, columns=["year","province","pop"])
```

```python
# New frame with missing column
frame2 = pd.DataFrame(data, columns=["year","province","pop","Birth rate"])
frame2
```

```python
type(frame["year"])
```

```python
# Retrieve a column with dict-syntax
frame2["province"]
```

```python
# Retrieve a column with dot-syntax
frame2.province
```

```python
# Assignment
frame2["Birth rate"] = 1.4
frame2
```

```python
# Assignment with an array of values
frame2["Birth rate"] = np.linspace(start=0.5, stop=2, num=6)
frame2
```

```python
# Assignment aligned by index
val = pd.Series([-1.2, -1.5, -1.7], index=[2, "four", "five"])
#val
frame2["Birth rate"] = val
frame2
```

```python
#print(frame2)
#frame2["pop"] > 4
frame2["is_western"] = frame2.province == "Alberta"
frame2
```
